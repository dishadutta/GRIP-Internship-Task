**Importing all the required libraries**

```
import sklearn.datasets as datasets
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
from google.colab import files
from sklearn.model_selection import train_test_split
!pip install pydotplus
!apt-get install graphviz -y
```

**Loading the dataset**

```
iris=datasets.load_iris()
dt=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```
dt.head()
```

**Data Inspection**

```
# concise summary of the DataFrame
dt.info()
```

```
# set of field definitions and related metadata
dt.shape
```

```
#Summary statistics of the iris data
dt.describe(include = 'all')
```

```
# checks if there is any missing value in the dataset
dt.isnull().sum()
```

```
y=iris.target
print(y)
```

**Splitting the dataset into train sets and test sets**

```
X_train, x_test, y_train, y_test = train_test_split(dt, y, test_size=0.33, random_state=42)
```

**Defining the Decision Tree Algorithm**

```
# Defining the decision tree algorithm
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)

print('Decision Tree Classifer Created')
```

**Visualizing the Decision Tree to understand it better**



```
# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

**Hence, now we can feed any new or test data to this classifer and it would be able to predict the right class accordingly.**
